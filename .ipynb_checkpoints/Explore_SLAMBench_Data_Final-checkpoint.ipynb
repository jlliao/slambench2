```python
## For example to generate the pareto Accuracy / ExecutionTime

import pandas as pd
import matplotlib.pyplot as plt 
import re
#'MaxATE':  x ['Summary']['unnamed']['AbsoluteError']['MAX'], 'FPS':  (1 / x ['Summary']['algo']['Duration_Frame']['MEAN'])

mycleandata1 = [ {'x': float(re.search('ICL2-KF-INT-(.+?).log', k).group(1)), 'MaxATE':  x ['Summary']['unnamed']['AbsoluteError']['MAX']} for (k,x) in list(data_icl2_kf_int.items())[:] ]
mycleandata2 = [ {'x': float(re.search('TUM1XYZ-KF-INT-(.+?).log', k).group(1)), 'MaxATE':  x ['Summary']['unnamed']['AbsoluteError']['MAX']} for (k,x) in list(data_tum1_kf_int.items())[:] ]
mycleandata3 = [ {'x': float(re.search('TUM2XYZ-KF-INT-(.+?).log', k).group(1)), 'MaxATE':  x ['Summary']['unnamed']['AbsoluteError']['MAX']} for (k,x) in list(data_tum2_kf_int.items())[:] ]

config = [0.00001, 0.00002, 0.00003, 0.00004, 0.00005, 0.00006, 0.00007, 0.00008, 0.00009, 0.0001]
kfdf = pd.DataFrame()
orbdf = pd.DataFrame()
infdf = pd.DataFrame()
kfdf['config'] = config
orbdf['config'] = config
infdf['config'] = config
df1 = pd.DataFrame(mycleandata1).sort_values('x')
df2 = pd.DataFrame(mycleandata2).sort_values('x')
df3 = pd.DataFrame(mycleandata3).sort_values('x')
df = df1
df=df.rename(columns = {'MaxATE':'MaxATE1'})
df['MaxATE2'] = df2['MaxATE']
df['MaxATE3'] = df3['MaxATE']
df['avg'] = df[['MaxATE1', 'MaxATE2', 'MaxATE3']].mean(axis=1)
kfdf['avgint']=df['avg']

plt.plot( 'x', 'MaxATE1', data=df, marker='o', color='blue', linestyle='dashed', linewidth=2, label="ICL2")
plt.plot( 'x', 'MaxATE2', data=df, marker='o', color='red', linestyle='dashed', linewidth=2, label="TUM1")
plt.plot( 'x', 'MaxATE3', data=df, marker='o', color='green', linewidth=2, linestyle='dashed', label="TUM2")
plt.plot( 'x', 'avg', data=df, marker='o', color='black', linewidth=4, label="Average")
plt.ylim(0.01, 0.5)
plt.legend()

# plt.scatter(df["configuration"], df["MaxATE"])
# plt.xlim(0, 0.0001)
# plt.show()
```
